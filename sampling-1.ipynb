```
def sampling_systematic(dataset, samplings):
  interval = len(dataset) // samplings
  random.seed(1)
  begin = random.randint(0, interval)
  indexes = np.arange(begin, len(dataset), step = interval)
  samplings_systematic = dataset.iloc[indexes]
  return samplings_systematic
```

```
df_sampling_systematic = sampling_systematic(dataset, 100)
df_sampling_systematic.shape
```

```
def sampling_cluster(dataset, group_number):
  interval = len(dataset) / group_number
  groups = []
  group_id = 0
  count = 0
  for _ in dataset.iterrows():
    groups.append(group_id)
    count += 1
    if count > interval:
      count = 0
      group_id += 1
  dataset['group'] = groups
  random.seed(1)
  selected_group = random.randint(0, group_number)
  return dataset[dataset['group'] == selected_group]
```
