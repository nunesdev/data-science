```
# contar os dias (separados por vírgula), apenas mean() não funcionou
days_count = df_membros['days_per_week'].dropna().apply(lambda x: len(x.split(',')))

# Número médio de dias na semana que frequentam
media_days_per_week = days_count.mean()
print(media_days_per_week)

# Número médio de visitas por semana
media_visit_per_week = df_membros['visit_per_week'].mean()
print(media_visit_per_week)

# Desvio padrão da idade
std_idade = df_membros['Age'].std()
print(std_idade)

# Quantos tipos diferentes de bebida existem
tipos_bebidas = df_membros['fav_drink'].dropna().nunique()
print(tipos_bebidas)

# Grupo de aula mais frequentado
grupo_aula_mais_frequentado = df_membros['fav_group_lesson'].mode().iloc[0]
print(grupo_aula_mais_frequentado)
```

```
df_resumo = df_membros[[
    'Age',               # idade
    'abonoment_type',    # tipo de plano
    'visit_per_week',    # visitas por semana
    'days_per_week',     # dias por semana
    'avg_time_in_gym',   # tempo médio na academia
    'attend_group_lesson', # participa de aulas em grupo
    'personal_training', # uso de personal trainer
    'uses_sauna'         # uso da sauna
]].copy()

df_resumo.head()
```


- **Age** → permite segmentar clientes por faixa etária e identificar padrões de frequência.
- **abonoment_type** → ajuda a analisar a relação entre tipo de plano e comportamento do cliente.  
- **visit_per_week** e **days_per_week** → medem a frequência de uso da academia, fundamentais para identificar clientes mais ativos ou em risco de cancelamento.  
- **avg_time_in_gym** → mostra o tempo médio de permanência, que pode indicar nível de engajamento.  
- **attend_group_lesson** → revela adesão a aulas coletivas, importante para entender preferências e criar programas direcionados.  
- **personal_training** → identifica clientes que usam serviços adicionais, potencializando estratégias de venda.  
- **uses_sauna** → indica uso de facilidades extras, útil para avaliar retorno sobre investimentos em infraestrutura.  

 **Possíveis insights com essas colunas:**
- Identificar perfis de clientes com maior engajamento.  
- Avaliar se o tipo de plano influencia na frequência e no tempo de permanência.  
- Detectar faixas etárias mais propensas a contratar personal trainers ou participar de aulas coletivas.  
- Entender a correlação entre uso de serviços extras (como sauna) e retenção de clientes.
