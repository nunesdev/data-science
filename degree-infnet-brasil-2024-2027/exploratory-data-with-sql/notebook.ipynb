```
df_2 = _dntk.execute_sql(
  '-- Analisa a distribuição das notas de matemática.\n-- Cria faixas manuais (0–40, 41–70, 71–100) e conta quantos alunos caem em cada faixa.\n\nSELECT \n    CASE\n        WHEN "math score" BETWEEN 0 AND 40 THEN \'0–40\'\n        WHEN "math score" BETWEEN 41 AND 70 THEN \'41–70\'\n        WHEN "math score" BETWEEN 71 AND 100 THEN \'71–100\'\n    END AS faixa_matematica,\n    COUNT(*) AS total_alunos\nFROM "StudentsPerformance.csv"\nGROUP BY faixa_matematica\nORDER BY faixa_matematica;\n',
  'SQL_DEEPNOTE_DATAFRAME_SQL',
  audit_sql_comment='',
  sql_cache_mode='cache_disabled',
  return_variable_type='dataframe'
)
df_2
```

```
df_3 = _dntk.execute_sql(
  '-- Agrega as notas de leitura e escrita por faixa_matematica.\n-- Mostra como o desempenho em matemática se relaciona com outras disciplinas.\n\nSELECT \n    CASE\n        WHEN "math score" BETWEEN 0 AND 40 THEN \'0–40\'\n        WHEN "math score" BETWEEN 41 AND 70 THEN \'41–70\'\n        WHEN "math score" BETWEEN 71 AND 100 THEN \'71–100\'\n    END AS faixa_matematica,\n    AVG("reading score") AS media_leitura,\n    AVG("writing score") AS media_escrita\nFROM "StudentsPerformance.csv"\nGROUP BY faixa_matematica\nORDER BY faixa_matematica;\n',
  'SQL_DEEPNOTE_DATAFRAME_SQL',
  audit_sql_comment='',
  sql_cache_mode='cache_disabled',
  return_variable_type='dataframe'
)
df_3
```

```
df_4 = _dntk.execute_sql(
  '-- Cria coluna faixa_leitura para classificar o desempenho em leitura.\n\nSELECT \n    CASE\n        WHEN "reading score" BETWEEN 0 AND 50 THEN \'0–50\'\n        WHEN "reading score" BETWEEN 51 AND 80 THEN \'51–80\'\n        WHEN "reading score" BETWEEN 81 AND 100 THEN \'81–100\'\n    END AS faixa_leitura,\n    COUNT(*) AS total_alunos\nFROM "StudentsPerformance.csv"\nGROUP BY faixa_leitura\nORDER BY faixa_leitura;\n',
  'SQL_DEEPNOTE_DATAFRAME_SQL',
  audit_sql_comment='',
  sql_cache_mode='cache_disabled',
  return_variable_type='dataframe'
)
df_4
```

```
df_5 = _dntk.execute_sql(
  '-- Distribui os alunos por gênero e faixa de leitura.\n-- Permite analisar diferenças de desempenho em leitura entre meninos e meninas.\n\nSELECT \n    gender,\n    CASE\n        WHEN "reading score" BETWEEN 0 AND 50 THEN \'0–50\'\n        WHEN "reading score" BETWEEN 51 AND 80 THEN \'51–80\'\n        WHEN "reading score" BETWEEN 81 AND 100 THEN \'81–100\'\n    END AS faixa_leitura,\n    COUNT(*) AS total_alunos\nFROM "StudentsPerformance.csv"\nGROUP BY gender, faixa_leitura\nORDER BY gender, faixa_leitura;',
  'SQL_DEEPNOTE_DATAFRAME_SQL',
  audit_sql_comment='',
  sql_cache_mode='cache_disabled',
  return_variable_type='dataframe'
)
df_5
```

```
df_6 = _dntk.execute_sql(
  '-- Verifica a diversidade de categorias em duas colunas demográficas.\n-- COUNT(DISTINCT ...) retorna o número de níveis únicos.\n\nSELECT\n    COUNT(DISTINCT "parental level of education") AS niveis_escolaridade_pais,\n    COUNT(DISTINCT "race/ethnicity") AS niveis_etnia\nFROM "StudentsPerformance.csv";\n',
  'SQL_DEEPNOTE_DATAFRAME_SQL',
  audit_sql_comment='',
  sql_cache_mode='cache_disabled',
  return_variable_type='dataframe'
)
df_6
```

```
df_7 = _dntk.execute_sql(
  '-- Cria acoluna faixa_escrita para classificar o desempenho em escrita.\n-- Distribui os alunos em três intervalos de pontuação.\n\nSELECT \n    CASE\n        WHEN "writing score" BETWEEN 0 AND 50 THEN \'0–50\'\n        WHEN "writing score" BETWEEN 51 AND 80 THEN \'51–80\'\n        WHEN "writing score" BETWEEN 81 AND 100 THEN \'81–100\'\n    END AS faixa_escrita,\n    COUNT(*) AS total_alunos\nFROM "StudentsPerformance.csv"\nGROUP BY faixa_escrita\nORDER BY faixa_escrita;\n',
  'SQL_DEEPNOTE_DATAFRAME_SQL',
  audit_sql_comment='',
  sql_cache_mode='cache_disabled',
  return_variable_type='dataframe'
)
df_7
```

```
df_8 = _dntk.execute_sql(
  '-- Analisa o impacto do curso de preparação.\n-- Para cada grupo (completed / none), calcula:\n-- - Média das três notas\n-- - Amplitude (máximo - mínimo) das três notas\n\nSELECT\n    "test preparation course",\n    AVG("math score") AS media_matematica,\n    MAX("math score") - MIN("math score") AS amplitude_matematica,\n    AVG("reading score") AS media_leitura,\n    MAX("reading score") - MIN("reading score") AS amplitude_leitura,\n    AVG("writing score") AS media_escrita,\n    MAX("writing score") - MIN("writing score") AS amplitude_escrita\nFROM "StudentsPerformance.csv"\nGROUP BY "test preparation course"\nORDER BY "test preparation course";\n',
  'SQL_DEEPNOTE_DATAFRAME_SQL',
  audit_sql_comment='',
  sql_cache_mode='cache_disabled',
  return_variable_type='dataframe'
)
df_8
```
